```python
import pandas as pd

# URL file Iris.csv
url_iris = 'https://raw.githubusercontent.com/boedybios/kaggle_explorations/master/Iris_Data_Classification_and_EDA/dataset/iris/Iris.csv'

# URL file titanicfull.csv
url_titanic = 'https://raw.githubusercontent.com/boedybios/kaggle_explorations/master/belajar_matplotlib_dasar/data/titanicfull.csv'

# Membaca file CSV dari URL
df_iris = pd.read_csv(url_iris)
df_titanic = pd.read_csv(url_titanic)

# Menampilkan 5 baris pertama dari data Iris
print("Data Iris:")
print(df_iris.head())

# Menampilkan 5 baris pertama dari data Titanic
print("\nData Titanic:")
print(df_titanic.head())
```

```python
df = pd.read_csv('./data/titanicfull.csv')
df.columns = ['Pclass', 'Survival status', 'full Name', 'Sex  ', '  Age', 
              'Sib SP', 'Parch', 'Ticket', 'Fare', 'Cabin', 'Embarked']
df_backup = df.copy(deep=True)

df.head()
```

```python
#Menggunakan lowercase untuk nama kolom dan mengganti spasi dengan _
df.columns = df.columns.str.replace(' ', '_').str.lower()
df.head()
```

```python
#Memangkas kelebihan spasi pada nama kolom
df = df_backup.copy(deep=True)

df.columns = df.columns.str.lower().str.strip().str.replace(' ', '_')
df.head()
```

```python
#11: Melakukan seleksi kolom dan baris pada Data Frame menggunakan loc
import pandas as pd
import numpy as np

print(pd.__version__)
print(np.__version__)
```

```python
n_rows = 10
n_cols = 5
cols = tuple('ABCDE')

df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)), 
                  columns=cols)
df
```

```python
#Seleksi kolom dan baris menggunakan loc
df.loc[[0,3,4], ['B','E']]
```

```python
#Seleksi baris dengan kondisi
df.loc[df['B']>10, ['B','D','E']]
```

```python
#Slicing Data Frame dengan loc
df.loc[0:4, 'B':'D']
```

```python
#12: Membentuk kolom bertipe datetime dari sejumlah kolom lain pada Data Frame
import pandas as pd

print(pd.__version__)
```

```python
data = {'day':[1, 2, 10 ,25, 12], 
        'month':[1, 2, 4, 5, 6], 
        'year':[2000, 2001, 2010, 2015, 2020]}

df = pd.DataFrame(data)
df
```

```python
#Membentuk kolom bertipe datetime
df['penaggalan'] = pd.to_datetime(df[['day', 'month', 'year']])
df
```
